```python
import ee, eemont, geemap
from GEE_ImageFusion import *
import main
import time

ee.Initialize()
ee.Authenticate()
```

```python
def run_data_fusion_for_year(year):
    startDate = f'{year}-01-01'
    endDate = f'{year}-12-31'
    task = main.data_fusion(startDate, endDate)
    return task

# List of years for which to run the data fusion
years = range(2013, 2024)  # Example range from 2013 to 2023

tasks = []
for year in years:
    task = run_data_fusion_for_year(year)
    task.start()
    tasks.append((year, task))

# Function to monitor the status of the tasks
def check_task_status(tasks):
    for year, task in tasks:
        status = task.status()
        print(f"Year: {year}, Task ID: {task.id}, Status: {status['state']}")

# Check the status of all tasks
check_task_status(tasks)
```
